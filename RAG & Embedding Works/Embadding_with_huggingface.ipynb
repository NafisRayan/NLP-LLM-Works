```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Step 1: Load your embedding model
# embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")

# Step 2: Load and split the document content into smaller chunks (e.g., sentences or paragraphs)
with open('file.txt', 'r') as file:
    file_content = file.read()

# Split the document into chunks (e.g., sentences)
chunks = file_content.split('. ')  # Simple sentence splitting; adjust as needed

# Step 3: Compute embeddings for each chunk
chunk_embeddings = [embed_model.get_text_embedding(chunk) for chunk in chunks]

```

```python
# Step 4: Encode the question
question = "most expencive house in berlin?"
question_embedding = embed_model.get_text_embedding(question)

# Step 5: Compare the question embedding with each chunk embedding using cosine similarity
similarity_scores = cosine_similarity([question_embedding], chunk_embeddings)

# Step 6: Find the chunk with the highest similarity score
best_chunk_index = np.argmax(similarity_scores)
best_chunk = chunks[best_chunk_index]
best_score = similarity_scores[0][best_chunk_index]

# Step 7: Return the most relevant chunk as the answer
print(f"Similarity Score: {best_score}")
print(f"Answer: {best_chunk}")
```
